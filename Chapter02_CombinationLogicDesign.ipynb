#### Exercises 2.30

If only one input changes, the circuit does not have the potential to glitch. This is because: 
1. Every path from A to Y have 1 AND gate and 1 OR gate, therefore changes from A to Y should propagate at the same speed. 
2. B, C and D can only travel to Y through one path. Therefore, there cannot be asynchronous concurrent calculations.
