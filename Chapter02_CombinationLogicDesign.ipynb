#### Exercises 2.5
2.1 (a)<br>
Can be simplified by inspection: <br>
$\bar{A}B$

2.1 (b)<br>
Cannot be simplified.

2.1 (c)
Can be simplified to 2 implicant using Karnaugh maps: <br>
$AC + \bar{A}\bar{C} + A\bar{B}$

2.1 (d) <br>
Can be simplified to 3 implicant using Karnaugh maps: <br>
$\bar{B}\bar{D} + \bar{B}\bar{A} + AC\bar{D}$ <br>


2.1 (e) <br> 
Cannot be simplified. The Karnaugh map results in a diagonal pattern of 1s and 0s. There cannot be square larger than 1x1.
