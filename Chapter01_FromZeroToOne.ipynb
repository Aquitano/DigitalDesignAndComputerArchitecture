(a) <b> Hierarchy</b> can be used to split the car design into different systems (e.g. powertrain, chassis , suspension) and finer subsystems (e.g. powertrain can be decomposed into the engine, transmission, clutch, etc). <b>Modularity</b> is used to define the function and interface between the subsystems. For instance, the chassis mounts everything together, the powertrain provides tractions and the suspension dampens vertical impacts. <b>Regularity</b> can be used to uniformize the subsystems. E.g. different car models may use the same dampers models or the same chassis fasteners. 


#### Exercises 1.3
<b>Hierarchy</b> can be used to split the houses into sections, e.g. : 1. foundation, 2. framing, 3. walls, 4. roof, etc. <br>
<b>Modularity</b> can be used to infer function to each sections. The lower foundation must be rigid and support framing, the framing must support walls and allow for wires/plumbing routing, walls must be soundproof and the roof water resistant. <br>
<b>Regularity</b> is used to simplify construction. Every section is made of a small but different set of available material (e.g. foundation requires concrete, framing requires wood/nails, etc). Also every section can be made of a repeated pattern, e.g. the same roof pattern extends all across the top of the house.
